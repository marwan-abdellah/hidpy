```python
# Create a sample image 

import os 
import seaborn as sns
from matplotlib import pyplot


sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)

# Save the image 
pyplot.savefig('sample.png')
```

```python
# Load the HTML file 
html_file_path = '%s/../templates/index.html' % os.getcwd()

# Search for the image within the HTML file  
with open(html_file_path, 'r') as file:
    data = file.read().replace('TRAJECTORY_FILE', 'sample.png')

# Create the new image with the HTML file 
report_file = open('report.html', 'w')
report_file.write(data)
report_file.close()
```
